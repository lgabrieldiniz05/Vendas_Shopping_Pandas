```python
import pandas as pd
tabela_vendas = pd.read_excel('Vendas.xlsx')
#Imprimir arquivo na tela
display(tabela_vendas)
```

```python
#Filtrando por loja
vendas_norteshopping = tabela_vendas.loc[tabela_vendas['ID Loja'] == 'Norte Shopping']
display(vendas_norteshopping)
```

```python
#filtrando linhas e colunas
vendas_norteshipping_dois = tabela_vendas.loc[tabela_vendas['ID Loja'] == 'Norte Shopping', 
['ID Loja', 'Produto','Valor Unitário']]
display(vendas_norteshipping_dois)
```

```python
#filtrar por coluna
produto_valor = tabela_vendas[['Produto', 'Valor Unitário']]
display(produto_valor)
```
